```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
slices =[90 ,20 ]
labels =['ninety' , 'twenty']
color = ['blue' , 'red' ]
plt.pie(slices, labels = labels)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
slices =[90 ,20]
labels =['Sixty' , 'Fourty']
plt.pie(slices, labels = labels , wedgeprops = {'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
slices =[90 ,20 , 30 , 50 ]
labels =['ninety' , 'twenty' ,'thirty' , 'fifty']
color = ['blue' , 'red' , 'yellow' , 'green'  ]
plt.pie(slices, labels = labels)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Language Popularity
slices1 = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
plt.pie(slices1 , labels= labels , wedgeprops ={'edgecolor' :'black'})
```

```python
# Language Popularity
slices2 = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
plt.pie(slices2 , labels= labels , wedgeprops ={'edgecolor' :'black'})
```

```python
explode=[0 ,0 ,0 ,0.2 ,0]
plt.pie(slices2 , labels= labels , explode=explode,shadow=True , wedgeprops ={'edgecolor' :'black'})
```

```python
explode=[0 ,0 ,0 ,0.1 ,0]
plt.pie(slices2 , labels= labels , explode=explode,shadow=True , startangle=90, wedgeprops ={'edgecolor' :'black'})
```

```python
explode=[0 ,0 ,0 ,0.1 ,0]
plt.pie(slices2 , labels= labels , explode=explode,shadow=True , startangle=90, autopct='%1.1f%%' ,wedgeprops ={'edgecolor' :'black'})
```
