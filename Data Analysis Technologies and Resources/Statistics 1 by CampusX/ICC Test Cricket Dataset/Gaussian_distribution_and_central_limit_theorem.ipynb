#  Gaussian distribution and central limit theorem 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('ICC Test Bat 3001.xlsx')
data.tail()
```

```python
data = data.replace(to_replace='-', value=np.nan)
```

```python
data.isnull().sum()
```

```python
data = data.dropna(how='any')
data.isnull().sum()
```

```python
data
```

```python
sns.distplot(data['Runs'])
```

```python
runs_data_series = data['Runs']
runs_array= np.array(runs_data_series)
runs_array
```

```python
np.random.choice(runs_array  , size =5)
```

```python
number_of_samples = 50
number_of_batches = 1000

list_of_batches =[]

for i in range(number_of_batches):
    x=np.random.choice(runs_array , size=number_of_samples)
    list_of_batches.append(x.mean())
    
list_of_batches
```

```python
sns.distplot(list_of_batches)
```
