```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('iris.csv')
data.head()
```

```python
classes = data['variety'].value_counts()
classes
```

```python
data = data.rename(columns = {'sepal.length':'SL' , 'sepal.width':'SW' ,'petal.length':'PL' ,'petal.width':'PW'})
data
```

```python
iris_setosa = data[data['variety']=='Setosa']
iris_virginica = data[data['variety']=='Virginica']
iris_versicolor = data[data['variety'] == 'Versicolor']
```

```python
iris_setosa
```

# 1.Mean


```python
# Sepal Width

print(np.mean(iris_setosa['SW']))
print(np.mean(iris_versicolor['SW']))
print(np.mean(iris_virginica['SW']))
```

```python
# Sepal Length
print(np.mean(iris_setosa['SL']))
print(np.mean(iris_versicolor['SL']))
print(np.mean(iris_virginica['SL']))
```

```python
# Petal Width
print(np.mean(iris_setosa['PW']))
print(np.mean(iris_versicolor['PW']))
print(np.mean(iris_virginica['PW']))
```

```python
# Petal Length
print(np.mean(iris_setosa['PL']))
print(np.mean(iris_versicolor['PL']))
print(np.mean(iris_virginica['PL']))
```

## The problem of Outliers

```python
plt.hist(iris_setosa['PL'] , color='red' , edgecolor='black')
```

```python
# If we add an outlier of 50 at the end
np.mean(np.append(iris_setosa['PL'],50))
```

- In above cell we have added a outlier 50 in the dataset series. Because of this series the peltal length of iris setosa is increased by 1.4620000000000002 to 2.4137254901960787 
- We need to avoid using mean when there is outliers present in the dataset.
- Try to use all three measure of center tendency

# Median 
### Median is a middle term in a sorted array

```python
# Petal Length
print(np.median(iris_setosa['PL']))
print(np.median(iris_versicolor['PL']))
print(np.median(iris_virginica['PL']))
```

### The good thing about median is we can use it if we have outlier count less than 45% in dataset
### But to apply the median on dataset we need a sorted array or pandas series

```python
np.median(np.append(iris_setosa['PL'],50))
```

```python
arr= np.array([1,4,5,8,12,600,700,800,100000])
np.median(arr)
```

```python
array_setosa_PL = np.array([iris_setosa['PL']])
array_setosa_PL.sort()
```

```python
np.append(array_setosa_PL, 50)
```

```python
# As you can see here ... outliers does'nt affect on the above array 
print(np.median(array_setosa_PL))
```

# Mode
#### Mode is the value which appears most often in a set of dataset


 ###                              Mode = 6

```python
# We don't have any function in numpy to calculate the mode 
# but in scipy.stats library we have mode fucntion in built

from scipy import stats
```

```python
print(stats.mode(iris_setosa['PL']))
print(stats.mode(iris_versicolor['PL']))
print(stats.mode(iris_virginica['PL']))
```

### Which one should you use?
#### We should use all 3 to have a fair idea of the centrality of the data
