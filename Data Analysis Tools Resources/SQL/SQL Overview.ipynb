## Types of SQL Statements
#### DDL ( Data Definition Language)|
#### DCL (Data Control Language)
#### DML (Data Manipulation Language)
#### TCL (Transaction Control Language)

### DDL ( Data Definition Language)
<b>Create</b> - Use to create new database object<br>
<b>ALTER</b>- Modify existing database object<br>
<b>DROP</b> - Delete existing database objects<br>
<b>TRUNCATE</b> - use to remove all the rows from the table

### DCL (Data Control Language)
<b>GRANT</b> - Provide access rights on database<br>
<b>REVOKE</b> - Withdraw access rights on database 

### DML (Data Manipulation Language)
<b> INSERT</b> - used to create new row in table <br> 
<b> UPDATE</b> - Modify data in tables <br> 
<b> DELETE</b> - Delete Data from tables <br> 
<b> SELECT</b> - Retrieve Data from tables
