j] IN/NOT IN

here in the below command we can filter and get the movies having desired genre but we need to add multiple conditional statements in order to filter the data 
SELECT title FROM movies WHERE genre LIKE 'comedy' OR genre LIKE 'Action'

To replace the repetative work we can use IN or NOT IN method/statements 
SELECT title, genre FROM movies WHERE genre IN ("Action","Horror" ,"Drama")

SELECT title, genre FROM movies WHERE genre NOT IN ("Action","Horror" ,"Drama")


k] Wild Cards 
Wild cards are characters which you can use to search in a string

Below code is used to search the string starting from letter A
SELECT title FROM movies WHERE title LIKE 'A%'

Below code is used to search man in a string 
SELECT title FROM movie WHERE title LIKE '%man%'


Below code is used to search the string ending from letter A
SELECT title FROM movie WHERE title LIKE '%a'


SELECT title actor FROM movies WHERE Actor LIKE "%Khan% or Actor LIKE "%Kapoor%"


( _ ):
This below command will find the name of movies having 5 Charachter length
SELECT title FROM movies WHERE title LIKE '_ _ _ _ _ '
