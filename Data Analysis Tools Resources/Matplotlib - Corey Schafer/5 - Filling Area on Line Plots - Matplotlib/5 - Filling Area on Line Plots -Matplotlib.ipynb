```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
ages = data['Age']
dev_salaries = data ['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot( ages , dev_salaries , color ="#444444" , linestyle="--" , label="All Devs")
plt.plot(ages , py_salaries , label ='Python')
plt.fill_between(ages , py_salaries , alpha= 0.3)
plt.legend()
plt.title("Median Salaries (USD) by ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot( ages , dev_salaries , color ="#444444" , linestyle="--" , label="All Devs")
plt.plot(ages , py_salaries , label ='Python')
overall_median = 57287
plt.fill_between(ages , py_salaries ,overall_median, alpha= 0.25)
plt.legend()
plt.title("Median Salaries (USD) by ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot( ages , dev_salaries , color ="#444444" , linestyle="--" , label="All Devs")
plt.plot(ages , py_salaries , label ='Python')
overall_median = 57287
plt.fill_between(ages , py_salaries ,overall_median,
                where =(py_salaries >overall_median),
                interpolate=True, alpha= 0.25)
plt.legend()
plt.title("Median Salaries (USD) by ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot( ages , dev_salaries , color ="#444444" , linestyle="--" , label="All Devs")
plt.plot(ages , py_salaries , label ='Python')
overall_median = 57287
plt.fill_between(ages , py_salaries ,overall_median,
                where =(py_salaries >overall_median),
                interpolate=True, alpha= 0.25
                ,color='green')

plt.fill_between(ages , py_salaries ,overall_median,
                where =(py_salaries <=overall_median),
                interpolate=True, alpha= 0.25
                ,color="red") 
plt.legend()
plt.title("Median Salaries (USD) by ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot( ages , dev_salaries , color ="#444444" , linestyle="--" , label="All Devs")
plt.plot(ages , py_salaries , label ='Python')
overall_median = 57287
plt.fill_between(ages , py_salaries ,dev_salaries,
                where =(py_salaries >dev_salaries),
                interpolate=True, alpha= 0.25
                ,color='green' , label="Above AVG")

plt.fill_between(ages , py_salaries ,dev_salaries,
                where =(py_salaries <=dev_salaries),
                interpolate=True, alpha= 0.25
                ,color="red",label="Below AVG") 
plt.legend()
plt.title("Median Salaries (USD) by ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```
