## Video - 8

### How do I filter rows of a pandas DataFrame by column value?

```python
movies7 = pd.read_csv("https://bit.ly/imdbratings")
movies7.head()
```

```python
movies7.shape
```

```python
booleans = []
for movie in movies7.duration:
    if movie >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
        
```

```python
#  ----------- OR -------------
# To replace the above for loop we can use below code (Efficient Code)

is_long = movies7.duration >= 200
is_long.head()
```

```python
len(booleans)
```

```python
# Converting list ot pandas series
is_long = pd.Series(booleans)
is_long.head()
```

```python
is_long.value_counts()
```

```python
# Showing the duration value rows having duration more than 200 minutes
# movies7[is_long]

#  ----------- OR -------------
# To replace the above code we can use below code (Efficient Code)

movies7[ movies7.duration >= 200]
```

```python
# This will pull out the type of genre having movie duration more than 200
# Avoid this below mehod (Try to use loc function for this type of filtering)
movies7[ movies7.duration >= 200]['genre']
```

## Video - 9|

### How do I apply multiple filter criteria to a pandas DataFrame?

```python
movies9 = pd.read_csv("https://bit.ly/imdbratings")
movies9.head()
```

```python
movies9[movies9['duration']>=200]
```

#### There Are three Opertors in Pandas that you can use to apply multiple filters

AND - &
OR  - |
Not - ~

```python
# Using or operator for filtering
movies9[(movies9['duration']>=200) | (movies9["genre"]=="Crime")]
```

```python
# Using and operator for filtering
movies9[(movies9['duration']>=200) & (movies9["genre"]=="Crime")]
```

```python
# Using not operator for filtering
movies9[~(movies9['duration']>=200)]
```

## Video - 16

### How do  handle missing values in pandas ?

```python
ufo16 = pd.read_csv('http://bit.ly/uforeports')
ufo16.tail()
```

```python
ufo.isnull().tail()
```

```python
ufo.isnull().sum()
```

```python
ufo16[ufo16['City'].isnull()]
```

##### What to do with the missing values 
<font color='green'>
1) You can drop it by Specifying some condition usinig dropna function

OR

2)  You can fill with some logical value using fillna function

</font>

```python
# Drop the all missing values
```

```python
ufo.shape
```

```python
# This will drop full row if there is any one or more value are null
ufo.dropna(how='any').shape
```

```python
# This will drop full row if all values in a single row are null
ufo.dropna(how='any').shape
```

```python
# This will drop the row if City and Shape Reported column both values are missing
ufo.dropna(subset = ['City' , 'Shape Reported'] , how ='all').shape
```

```python
# Filling missing values ( Missing values are excluded because we haven't used inplace=True )
ufo['Shape Reported'].value_counts()
```

```python
# All the missing values are filled with the VARIOUS word
ufo['Shape Reported'].fillna(value = 'VARIOUS' , inplace=True)
```

```python
ufo['Shape Reported'].value_counts()
```
