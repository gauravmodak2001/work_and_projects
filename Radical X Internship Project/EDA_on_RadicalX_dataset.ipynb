### 2. Intenship_program_type	

```python
temp2 = df.Intenship_program_type.str.split(',')
# temp2.sample(20)
# temp2.unique()
```

```python
temp2_1 = []
for i in range(100): 
    for element in temp2[i]:
        if element in temp2_1:
            continue
        else:
            temp2_1.append(element)
print(temp2_1)
```

```python
count2_1 = {}

for i in temp2:
    for element in i:
        if not element in count2_1:
            count2_1[element] =1
        else:
            count2_1[element]+=1

# count2_1
```

```python
df2 = pd.Series(count2_1).sort_values(ascending=True)
df2 = df2.to_frame()
```

```python
plt.barh(df2.index , df2[0] , label='Intenship Program Type' , color='y')
plt.xlabel("Users Count")
plt.ylabel("Programming languages")
plt.legend()
plt.grid()
plt.show()
```
