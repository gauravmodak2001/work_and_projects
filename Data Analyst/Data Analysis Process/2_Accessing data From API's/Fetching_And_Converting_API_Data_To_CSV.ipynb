#### Making dataframe which include following columns
1. Popularity
2. Vote Count
3. Original Lang
4. Title 
5. vote avg
6. Overview
7. Release Date
   
   
10000*7  ==> shape

```python
# fectchig a single column from the API results
for i in data_json['results']:
    print(i['title'] , i['vote_average'])
```

```python
# Making a dataframe using above operations 
popularity= []
release_date=[]
title =[]
original_title =[]
vote_average= []
vote_count =[]
original_language=[]
overview =[]

for i in data_json['results']:
    popularity.append(i['popularity'])
    release_date.append(i['release_date'])
    title.append(i['title'])
    original_title.append(i['original_title'])
    vote_average.append(i['vote_average'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    overview.append(i['overview'])
    
df = pd.DataFrame({'overview':overview, 'title':title, 'original_title':original_title ,'popularity':popularity , 'release_date':release_date , 'vote_count':vote_count ,'vote_average':vote_average ,  "original_language":original_language})
```
