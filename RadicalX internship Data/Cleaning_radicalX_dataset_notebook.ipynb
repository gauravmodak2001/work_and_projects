```python
All_Languages_list1 = ['english' ,'Español', 'french' , 'arabic' ,'spanish' , 'español' ,'no' , 'hindi' , 'bangla' , 'bengali' ,'portuguese','Ukrainian' , 'russian' ,'punjabi' , 'mandarin' , 'japanese' , 'turkish' ,'vietnamese' ,'german' ,'telugu' , 'marathi' , 'tamil', 'malayalam' , 'gujarati' ,'bahasa' , 'Indonesia' ,'Belarusian' , 'Nepali' ,'Igbo' ,'Yoruba' , 'Hausa' ,'Slovenian' , 'Urdu' ,'Sindhi' ,'Polish' , 'Swahili' ,'Persian' , 'Tigrinya' ,  'Korean''Chinese' , ]
All_Languages_list=[]
for i in All_Languages_list1:
    i.lower()
    All_Languages_list.append(i)
    
communication_lang = df.preferred_communication_language.str.split(' ')
```

```python
new_final_list =[]

for i in communication_lang:
    temp = []
    for x in i:
        if x in temp:
            continue
        elif x in All_Languages_list:  
            temp.append(x)
#         print(x)

    new_final_list.append(temp)
new_final_list
```

```python
where_did_you_hear_about_this_internship_seperated = df['where_did_you_hear_about_this_internship?'].str.split(' ')
```

```python
unique_values_where_did_you_hear_about_this_internship = ['linkedin' , 'indeed' , 'web' ,'google' ,'whatsapp' ,'instagram' , 'twitter' , 'facebook' ,'telegram' ,'discord' ]

unique_values_where_did_you_hear_about_this_internship_final =[]

for i in where_did_you_hear_about_this_internship_seperated:
    temp= []
    for x in i:
        if x in unique_values_where_did_you_hear_about_this_internship:
            temp.append(x)
#             print(x)
            break
        else:
            temp.append('referral')
            break

    unique_values_where_did_you_hear_about_this_internship_final.append(temp)
unique_values_where_did_you_hear_about_this_internship_final
```
