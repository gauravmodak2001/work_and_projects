```python
iris_setosa=data[data['variety']=='Setosa']
iris_Virginica=data[data['variety']=='Virginica']
iris_Versicolor=data[data['variety']=='Versicolor']
```

# Cumulative Density Function (PDF)

```python
counts , bin_edges = np.histogram(iris_setosa['PL'] , bins=10)
```

```python
print(counts)
print("-------------------------------------------------------------")
print(bin_edges)
```

```python
pdf = counts /sum(counts)
cdf = np.cumsum(pdf)

print(pdf)
print(cdf)
```

```python
plt.plot(bin_edges[1:] , pdf , label='PDF curve')
plt.plot(bin_edges[1:] , cdf , label='CDF curve')
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
counts , bin_edges = np.histogram(iris_setosa['PL'] , bins=10)
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:] , pdf , label=' Setosa PDF curve')
plt.plot(bin_edges[1:] , cdf , label=' Setosa CDF curve')


counts , bin_edges = np.histogram(iris_Versicolor['PL'] , bins=10)
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:] , pdf , label='Versicolor PDF curve')
plt.plot(bin_edges[1:] , cdf , label='Versicolor CDF curve')


counts , bin_edges = np.histogram(iris_Virginica['PL'] , bins=10 )
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:] , pdf , label='Virginica PDF curve')
plt.plot(bin_edges[1:] , cdf , label='Virginica CDF curve')

plt.legend()
plt.show()
```
